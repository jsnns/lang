```python
import re 

base_path = "/Users/jacobsansbury/Projects/jsnns/journal/"
files = [
    "./2019/10/22.md",
    "./2019/09/18.md",
    "./2019/09/19.md",
    "./2019/09/27.md",
    "./2019/08/28.md",
    "./2019/08/29.md",
    "./2019/12/11.md",
    "./2021/01/01.md",
    "./2020/02/21.md",
    "./2020/02/25.md",
    "./2020/10/23.md",
    "./2020/01/08.md",
    "./2020/01/27.md",
    "./2018/09/13.md",
    "./2018/09/14.md",
    "./2018/09/05.md",
]
files = [file.replace("./", base_path) for file in files]
non_dec = re.compile(r'[^\d.]+')

def get_rows(path):
    with open(path) as file:
        content = [x.strip() for x in file.readlines()] 
        stats_index = None
        for index, line in enumerate(content):
            if "# Life Stats" in line:
                stats_index = index
    return content[stats_index+4:]

def parse_row(row):
    return [cell.strip() for cell in row if cell != ""]

def make_dict(row):
    return {key: value for (key, value) in row}

def remove_alpha_chars(s):
    return  re.sub(non_dec,'', s)

def normalize_data(d):
    if "Networth" in d:
        d["Networth"] = float(remove_alpha_chars(d["Networth"]))
        if d["Networth"] < 2000:
#             likely was represented with "k"
            d["Networth"] *= 1000
    return d

def parse_file(f):
    rows = get_rows(f)
    rows = [row.split("|") for row in rows]
    rows = [row for row in rows if len(row) > 1]
    return normalize_data(make_dict([parse_row(row) for row in rows]))

def parse_files(fs):
    return [parse_file(f) for f in fs]

parse_files(files)
```
